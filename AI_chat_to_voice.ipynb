```python
!pip install gtts

import os
from openai import OpenAI
from gtts import gTTS
import IPython.display as display
import time

# Securely load your API key from an environment variable
api_key = "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"
if not api_key:
    raise ValueError("API Key is required")

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=api_key,
)

conversation_history = []

print("You can start chatting with the bot. Type 'exit' to quit.")

def text_to_speech(text):
    """Converts text to speech and plays it."""
    tts = gTTS(text=text, lang='en', tld='com')
    tts.save("output.mp3")
    display.display(display.Audio("output.mp3", autoplay=True))
    time.sleep(len(text) / 10)

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Exiting the chat.")
        break

    conversation_history.append({"role": "user", "content": user_input})

    try:
        completion = client.chat.completions.create(
            model="cognitivecomputations/dolphin-mixtral-8x7b",
            messages=conversation_history
        )

        bot_reply = completion.choices[0].message.content
        print(f"Bot: {bot_reply}")
        conversation_history.append({"role": "assistant", "content": bot_reply})

        text_to_speech(bot_reply)

    except Exception as e:
        print(f"An error occurred: {e}")
```
